```python
import pandas as pd
```

```python
csv_file_path = '/Users/april/Downloads/popular_brand.csv'
df = pd.read_csv(csv_file_path,header=None)
```

```python
brands = df.iloc[:,0].tolist()
brands[1:10]
len(brands)
```

```python
brands = [x.lower() for x in brands]
brands
```

```python
from youtube_api import YouTubeDataAPI

api_key = ''
yt = YouTubeDataAPI(api_key)
```

```python
lst1 = yt.search('estee lauder')
len(lst1)
```

```python
brand_video = []

for b in brands:
    lst1 = yt.search(b)
    for dict in lst1:
        dict.update({'keyword': b})
    brand_video+=lst1
```

```python
len(brand_video)
```

```python
import csv
csv_file_path = '/Users/april/Downloads/by_brand.csv'

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = brand_video[0].keys()
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write header
    writer.writeheader()

    # Write data
    writer.writerows(brand_video)

print(f'Data has been successfully written to {csv_file_path}')
```

```python
csv_file_path2 = '/Users/april/Downloads/by_brand.csv'
brand_videos = pd.read_csv(csv_file_path2)
```

```python
brand_videos.info()
```

```python
video_ids = brand_videos['video_id'].tolist()
len(video_ids)
```

```python
import time
import random
import requests
```

```python
video_info = []
for id in video_ids:
    try:
        dict1 = yt.get_video_metadata(id)
        video_info.append(dict1)
        time.sleep(random.randint(1, 3)) 
        
        # invalid HTTP response: HTTP error
    except requests.exceptions.HTTPError as http_err:
        print("HTTP error:", http_err, "\n")
        requested_page_txt = str(http_err)  # str(http_err.args) default format of .args is tuple

    # failed connection: Connection Error
    except requests.exceptions.ConnectionError as connection_err:
        print ("Connection Error:", connection_err, "\n")
        requested_page_txt = str(connection_err)  
        
    # request timed out: The request timed out
    except requests.exceptions.Timeout as timeout_err:
        print ("Timeout Error:", timeout_err, "\n")
        requested_page_txt = str(timeout_err)  
       
    # generic exception for other errors
    except requests.exceptions.RequestException as other_err:
        print ("Unknown Error:", other_err, "\n")
        requested_page_txt = str(other_err)  


    
```

```python
len(video_info)
```

```python
csv_file_path = '/Users/april/Downloads/by_video_id.csv'

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = video_info[0].keys()
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write header
    writer.writeheader()

    # Write data
    writer.writerows(video_info)

print(f'Data has been successfully written to {csv_file_path}')
```
